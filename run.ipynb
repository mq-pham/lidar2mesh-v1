```python
import torch
import torch.nn.functional as F 


def rot6d_to_rotmat(x):
    x = x.view(-1,3,2)

    # Normalize the first vector
    b1 = F.normalize(x[:, :, 0], dim=1, eps=1e-6)

    dot_prod = torch.sum(b1 * x[:, :, 1], dim=1, keepdim=True)
    # Compute the second vector by finding the orthogonal complement to it
    b2 = F.normalize(x[:, :, 1] - dot_prod * b1, dim=-1, eps=1e-6)

    # Finish building the basis by taking the cross product
    b3 = torch.cross(b1, b2, dim=1)
    rot_mats = torch.stack([b1, b2, b3], dim=-1)

    return rot_mats
```

```python
import torch
import matplotlib.pyplot as plt
from P4D.smplx_model import SMPLX
from lidar import LiDAR
from dataset import load_and_split_all_datasets
from trainer import train_one_epoch, evaluate, save_checkpoint, test, load_latest_checkpoint
from loss import Loss, WarmupMultiStepLR
from p4gru import P4GRU, PointGRU
from utils import init_weights, optimize_memory
import gc

def run_simulation(params):
    # Set up device
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    
    # LiDAR parameters (constant for all simulations)
    lidar_params = {
        'elev': 0.0, 'dist': 6.0, 'image_size': (256, 192),
        'min_depth': 0.0, 'max_depth': 1.0, 'return_normals': False, 'subsample': 1024
    }
    
    # Initialize models
    smpl = SMPLX().to(device)
    lidar = LiDAR(**lidar_params, device=device)
    
    # Load and prepare data
    list_data_names = ["../data/CNRS/283/SW_B_1_stageii.npz"]
    train_dataset, val_dataset, test_dataset = load_and_split_all_datasets(
        list_data_names, device, smpl, lidar, params['batch_size'], 0.8, 0.1, 0.1,
        params['step'], params['augment'], params['opti_trans']
    )
    train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=1, shuffle=True)
    val_loader = torch.utils.data.DataLoader(val_dataset, batch_size=1, shuffle=True)
    
    optimize_memory()  # Clear memory before training
    
    # Model setup
    factor = 3 if params['_type'] == "3D" else 6
    output_dim = (factor*22)*params['batch_size']
    if params['opti_beta']:
        output_dim += 16*params['batch_size']
    if params['opti_trans']:
        output_dim += 3*params['batch_size']
    
    model = P4GRU(spatial_kernel_size=[params['radius'], params['nsamples']],
                  spatial_stride=params['spatial_stride'], emb_relu=params['emb_relu'],
                  dim=params['dim'], depth=params['depth'], heads=params['heads'],
                  dim_head=params['dim_head'], mlp_dim=params['mlp_dim'],
                  output_dim=output_dim).to(device)
    
    if params['init_weight']:
        model.apply(init_weights)
    
    n_layers = sum(1 for attr in dir(model) if isinstance(getattr(model, attr), PointGRU))
    
    model_name = (
        f"P4GRU_{params['_type']}_"
        f"lr{params['lr']}_"
        f"beta{params['opti_beta']}_"
        f"trans{params['opti_trans']}_"
        f"init_weights{params['init_weight']}_"
        f"GRUlayers{n_layers}_"
        f"pos_emb{params['pos_emb']}_"
        f"augment{params['augment']}_"
        f"step{params['step']}_"
        f"radius{params['radius']}_"
        f"nsamples{params['nsamples']}_"
        f"spatial_stride{params['spatial_stride']}_"
        f"emb_relu{params['emb_relu']}_"
        f"dim{params['dim']}_"
        f"depth{params['depth']}_"
        f"heads{params['heads']}_"
        f"dim_head{params['dim_head']}_"
        f"mlp_dim{params['mlp_dim']}_"
        f"bs_{params['batch_size']}"
    )
    
    # Loss and optimizer setup
    criterion = Loss(smplx=smpl, device=device, optimize_beta=params['opti_beta'],
                     optimize_trans=params['opti_trans'], factor=factor)
    optimizer = torch.optim.Adam(model.parameters(), lr=params['lr'], weight_decay=1e-4)
    
    lr_scheduler = WarmupMultiStepLR(optimizer, milestones=[10, 15], gamma=0.1,
                                     warmup_iters=5 * len(train_loader), warmup_factor=1e-5)
    
    # Training loop
    num_epochs = 20
    all_loss_train = {k: [] for k in ['geodesic_loss', 'joints_loss', 'verts_loss',
                                      'beta_loss', 'trans_loss', 'theta_smooth', 'total_loss']}
    all_loss_val = {k: [] for k in ['geodesic_loss', 'joints_loss', 'verts_loss',
                                    'beta_loss', 'trans_loss', 'theta_smooth', 'total_loss']}
    
    print(f"Starting training for {model_name}")
    for epoch in range(num_epochs):
        train_loss = train_one_epoch(model, optimizer, criterion, train_loader,
                                     epoch, all_loss_train, print_freq=5, visu=False)
        val_loss = evaluate(model, criterion, val_loader, all_loss_val, visu=False)
        
        print(f"Epoch {epoch + 1}/{num_epochs}, train loss: {train_loss:.4f}, val loss: {val_loss:.4f}")
        
        learning_rate = optimizer.param_groups[0]['lr']
        save_checkpoint(model, optimizer, epoch + 1, train_loader, val_loader,
                        train_loss, val_loss, all_loss_train, all_loss_val,
                        model_name, learning_rate, params['batch_size'])
        
        lr_scheduler.step()
    
    print(f"Finished training for {model_name}")
    optimize_memory()  # Clear memory after training
    
    # Load the best model checkpoint
    best_checkpoint, _, _, _ = load_latest_checkpoint(model, optimizer, model_name)
    if best_checkpoint is not None:
        model.load_state_dict(best_checkpoint['model_state_dict'])
        print(f"Loaded best model from epoch {best_checkpoint['epoch']} with validation loss {best_checkpoint['val_loss']:.4f}")
    else:
        print("No checkpoint found. Using the model from the last training epoch.")
        

    output_dir = f"./viz/{model_name}"
    os.makedirs(output_dir, exist_ok=True)
    test_loss = test(model, criterion, test_loader, output_dir, visu=True)
    print(f"Test loss for {model_name}: {test_loss:.4f}")
    
    optimize_memory()  # Clear memory after testing

# Define the parameter sets for different simulations
simulation_params = [
    # Original parameters
    {'lr': 0.1, 'batch_size': 5, 'step': 1, 'augment': 1, 'opti_trans': False, 'opti_beta': False,
     'init_weight': True, 'pos_emb': True, 'radius': 0.5, 'nsamples': 5, 'spatial_stride': 2,
     'emb_relu': True, 'dim': 256, 'depth': 4, 'heads': 3, 'dim_head': 128, 'mlp_dim': 256, '_type': "3D"},
    
    {'lr': 0.001, 'batch_size': 7, 'step': 1, 'augment': 1, 'opti_trans': True, 'opti_beta': True,
     'init_weight': True, 'pos_emb': True, 'radius': 0.7, 'nsamples': 7, 'spatial_stride': 1,
     'emb_relu': False, 'dim': 512, 'depth': 6, 'heads': 4, 'dim_head': 64, 'mlp_dim': 512, '_type': "3D"},
    
    {'lr': 0.005, 'batch_size': 10, 'step': 1, 'augment': 1, 'opti_trans': False, 'opti_beta': True,
     'init_weight': False, 'pos_emb': False, 'radius': 0.6, 'nsamples': 12, 'spatial_stride': 2,
     'emb_relu': True, 'dim': 384, 'depth': 5, 'heads': 6, 'dim_head': 96, 'mlp_dim': 384, '_type': "3D"},
    
    {'lr': 0.002, 'batch_size': 8, 'step': 2, 'augment': 2, 'opti_trans': True, 'opti_beta': False,
     'init_weight': True, 'pos_emb': True, 'radius': 0.55, 'nsamples': 6, 'spatial_stride': 1,
     'emb_relu': True, 'dim': 320, 'depth': 5, 'heads': 5, 'dim_head': 64, 'mlp_dim': 320, '_type': "3D"},
    
    {'lr': 0.08, 'batch_size': 6, 'step': 1, 'augment': 1, 'opti_trans': False, 'opti_beta': True,
     'init_weight': False, 'pos_emb': True, 'radius': 0.65, 'nsamples': 8, 'spatial_stride': 2,
     'emb_relu': False, 'dim': 448, 'depth': 7, 'heads': 7, 'dim_head': 64, 'mlp_dim': 448, '_type': "3D"},
    
    {'lr': 0.003, 'batch_size': 9, 'step': 1, 'augment': 2, 'opti_trans': True, 'opti_beta': True,
     'init_weight': True, 'pos_emb': False, 'radius': 0.75, 'nsamples': 10, 'spatial_stride': 1,
     'emb_relu': True, 'dim': 288, 'depth': 6, 'heads': 4, 'dim_head': 72, 'mlp_dim': 288, '_type': "3D"},
    
    {'lr': 0.06, 'batch_size': 5, 'step': 2, 'augment': 1, 'opti_trans': False, 'opti_beta': False,
     'init_weight': False, 'pos_emb': True, 'radius': 0.45, 'nsamples': 4, 'spatial_stride': 2,
     'emb_relu': False, 'dim': 192, 'depth': 3, 'heads': 3, 'dim_head': 64, 'mlp_dim': 192, '_type': "3D"},
    
    {'lr': 0.005, 'batch_size': 12, 'step': 1, 'augment': 2, 'opti_trans': True, 'opti_beta': False,
     'init_weight': True, 'pos_emb': False, 'radius': 0.8, 'nsamples': 15, 'spatial_stride': 1,
     'emb_relu': True, 'dim': 576, 'depth': 8, 'heads': 9, 'dim_head': 64, 'mlp_dim': 576, '_type': "3D"},
    
    {'lr': 0.04, 'batch_size': 6, 'step': 2, 'augment': 1, 'opti_trans': False, 'opti_beta': True,
     'init_weight': True, 'pos_emb': True, 'radius': 0.7, 'nsamples': 32, 'spatial_stride': 2,
     'emb_relu': False, 'dim': 416, 'depth': 5, 'heads': 8, 'dim_head': 52, 'mlp_dim': 416, '_type': "3D"},
    
    {'lr': 0.07, 'batch_size': 8, 'step': 1, 'augment': 1, 'opti_trans': False, 'opti_beta': False,
     'init_weight': False, 'pos_emb': True, 'radius': 0.55, 'nsamples': 32, 'spatial_stride': 1,
     'emb_relu': True, 'dim': 352, 'depth': 4, 'heads': 4, 'dim_head': 88, 'mlp_dim': 352, '_type': "3D"}
]

# Run simulations
for i, params in enumerate(simulation_params):
    print(f"\nStarting simulation {i+1}/{len(simulation_params)}")
    run_simulation(params)
    gc.collect()  # Force garbage collection between simulations

print("All simulations completed.")
```
